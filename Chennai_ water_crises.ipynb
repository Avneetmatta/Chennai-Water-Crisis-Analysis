```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('chennai_reservoir_levels.csv')
df.head()
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'], format = '%d-%m-%Y')
```

```python
df.info()
```

```python
df['Year']=df['Date'].apply(lambda x:x.year)
```

```python
df.head()
```

```python
df['Month']=df['Date'].apply(lambda x:x.month)
```

```python
df.head()
```

```python
df.columns
```

```python
df['Total water level']=df[['POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM']].sum(axis=1)
```

```python
df.head()
```

```python
Yearwise =  df.groupby('Year', as_index = False).mean()
```

```python
Yearwise
```

```python
Monthwise = df.groupby('Month', as_index = False).mean()
```

```python
plt.figure(figsize = (10,9))
sns.barplot(x='Year', y='Total water level', data = Yearwise )
plt.title('Mean yearly of water level in four dams')
plt.show()
```

```python
plt.figure(figsize = (10,9))
sns.barplot(x='Month', y='Total water level', data = Monthwise )
plt.title('Mean Monthly of water level in four dams')
plt.show()
```

```python
plt.figure(figsize = (10,9))
Yearwise.plot(x='Year', y =['POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM'])
```
